```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data['SN'].unique())
total_players_df = pd.DataFrame([total_players], columns = ['Total Players'], index = ['Data Summary'])
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data['Item ID'].unique())
item_id = list(purchase_data['Item ID'])
price = list(purchase_data['Price'])
item_price = set()
for i in range(0,len(item_id)):
    new_tuple = (item_id[i], price[i])
    item_price.add(new_tuple)

sum = 0.0
for i in range(0,len(item_price)):
    current = item_price.pop()
    sum += current[1]

avg_price = round(sum/unique_items, 2)
avg_price = '${:,.2f}'.format(avg_price)

num_purchases = len(item_id)

total_rev = '${:,.2f}'.format(purchase_data['Price'].sum())
total_rev
```

```python
d = {'Number of Unique Items': unique_items, 'Average Price': avg_price, 'Number of Purchases': num_purchases, 'Total Revenue': total_rev}
purchase_analysis = pd.DataFrame(d, index=['Data Summary'])
purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python


genders = purchase_data['Gender']
sn = purchase_data['SN']

unique_gender_set = set()

for i in range(0,num_purchases):
    my_tuple = (sn[i], genders[i])
    unique_gender_set.add(my_tuple)

unique_gender = []
for i in range(0,total_players):
    my_tuple = unique_gender_set.pop()
    my_list = [my_tuple[0], my_tuple[1]]
    unique_gender.append(my_list)

unique_gender_df = pd.DataFrame(unique_gender, columns = ['Total Count', 'Gender'])
total_gender_df = unique_gender_df.groupby('Gender').count().sort_values('Total Count', ascending = False)
total_gender_df['Percentage of Players'] = total_gender_df['Total Count'].apply(lambda x: '{:.2%}'.format(x/total_players))
total_gender_df = total_gender_df.reset_index(drop = True)
total_gender_df.index= ['Male', 'Female', 'Others / Non-Disclosed']
total_gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_totals = purchase_data.groupby('Gender').count()
gender_totals
```

```python
gender_totals = purchase_data.groupby('Gender').count()
gender_totals = gender_totals.rename(columns = {'Purchase ID':'Purchase Count', 'SN': 'Average Purchase Price', 'Age': 'Total Purchase Value', 'Item ID':'Avg Total Purchase per Person'})
gender_totals.drop(['Item Name', 'Price'], axis = 1, inplace = True)
gender_totals['Total Purchase Value'] = purchase_data.groupby('Gender').sum()['Price'].apply(lambda x: '${:.2f}'.format(x))
gender_totals
```

```python
male_players = len(purchase_data['SN'].loc[purchase_data['Gender'] == 'Male'].unique())
female_players = len(purchase_data['SN'].loc[purchase_data['Gender'] == 'Female'].unique())
other_players = len(purchase_data['SN'].loc[purchase_data['Gender'] == 'Other / Non-Disclosed'].unique())

num_unique_players = [female_players, male_players, other_players]

avg_total_pp = []
for i in range(0,len(gender_totals['Total Purchase Value'])):
    total = gender_totals['Total Purchase Value'][i][1:]
    average = float(total)/num_unique_players[i]
    average = '{:.2f}'.format(average)
    avg_total_pp.append(average)

gender_totals['Avg Total Purchase per Person'] = avg_total_pp

gender_totals
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,150]
bins_label = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_data['Age Groups'] = pd.cut(purchase_data["Age"], bins, labels= bins_label, right = True)

age_df = purchase_data
age_df_unique = age_df.groupby(['Age Groups']).nunique().drop(columns = ['Purchase ID', 'Age', 'Item ID', 'Item Name', 'Price', 'Age Groups', 'Gender'])
age_df_unique.rename(columns = {'SN': 'Total Count'}, inplace = True)

age_average = []

for item in age_df_unique['Total Count']:
    average = float(item)/total_players
    average = '{:.2%}'.format(average)
    age_average.append(average)

age_df_unique['Percent of Players'] = age_average
age_df_unique
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
df_top_item1 = purchase_data.groupby(['Item ID', 'Item Name']).count().sort_values('Purchase ID', ascending = False).head()
df_top_item1 = df_top_item1.rename(columns = {'Purchase ID': 'Purchase Count'})
df_top_item1 = df_top_item1.drop(columns = ['SN', 'Age', 'Gender', 'Price', 'Age Groups'])
index_list = list(df_top_item1.index.values)

item_id_list = list(purchase_data['Item ID'])
item_name_list = list(purchase_data['Item Name'])
item_price_list = list(purchase_data['Price'])

my_set = set()

for i in range(1,len(item_id_list)):
    a = item_id_list[i]
    b = item_name_list[i]
    c = item_price_list[i]
    
    my_set.add((a,b,c))

my_item_price_list = []
for i in range(0,len(my_set)):
    element = my_set.pop()
    element = list(element)
    for item2 in index_list:
        if element[0] == item2[0]:
            my_item_price_list.append(element)
    
```

```python
df_top_item1
```

```python
trial = pd.DataFrame(my_item_price_list, columns = ['Item ID', 'Item Name', 'Price']).set_index(['Item ID', 'Item Name'])
trial
```

```python
new_df = pd.merge(df_top_item1, trial, how = 'left', on = ['Item ID', 'Item Name'])
new_df
df_top_item2 = purchase_data.groupby(['Item ID', 'Item Name']).sum().drop(columns = ['Purchase ID', 'Age'])
df_top_item2
final_df = pd.merge(new_df, df_top_item2, how = 'left', on = ['Item ID', 'Item Name']).rename(columns = {'Price_x': 'Item Price', 'Price_y': 'Total Purchase Value'})
final_df['Item Price'] = final_df['Item Price'].map('${:.2f}'.format)
final_df['Total Purchase Value'] = final_df['Total Purchase Value'].map('${:.2f}'.format)
final_df
```
